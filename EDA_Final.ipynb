## Изучим распределение женщин и мужчин

```python
female_df = df[df['sex'] == 'female']
male_df = df[df['sex'] == 'male']
categories = ['Телепередачи', 'Сериалы', 'Фильмы', 'Разное', 'Развлечения', 
              'Обучение', 'Интервью', 'Юмор', 'Видеоигры', 'Лайфстайл']

category_counts_fm = female_df[categories].sum().sort_values(ascending=False)
category_counts_m = male_df[categories].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(category_counts_fm.index, category_counts_fm.values, color='pink')

plt.title('Рейтинг категорий для женщин')
plt.xlabel('Категории')
plt.ylabel('Частота встречаемости')
plt.xticks(rotation=45, ha='right');
```

```python
plt.figure(figsize=(10, 6))
plt.bar(category_counts_m.index, category_counts_m.values, color='skyblue')

plt.title('Рейтинг категорий для мужчин')
plt.xlabel('Категории')
plt.ylabel('Частота встречаемости')
plt.xticks(rotation=45, ha='right');
```

```python
# Шаг 1: Группировка данных по регионам и подсчет мужчин и женщин
region_sex_counts = df.groupby(['region', 'sex']).size().unstack(fill_value=0)

# Шаг 2: Определение, в каком регионе больше мужчин и женщин
region_dominance = region_sex_counts.apply(lambda row: 'male' if row['male'] > row['female'] else 'female', axis=1)

# Шаг 3: Подсчет количества регионов, где больше мужчин и где больше женщин
dominance_counts = region_dominance.value_counts()

# Шаг 4: Построение пайчарта
plt.figure(figsize=(8, 6))
plt.pie(dominance_counts, labels=dominance_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'pink'])

# Настройки графика
plt.title('Соотношение регионов с преобладанием мужчин и женщин')
plt.axis('equal')  # Делает круг ровным

# Показать график
plt.show()
```

## Выводы

- Большая часть пользователей использует смартфоны
- Фильмы, сериалы и телепередачи имеют распределение возрастов сходное, со всем датасетом. Особо выделяются категории интервью и разное - основная аудитория 35+, лайфстайл и юмор - 20 - 40 лет. Видим, что некоторые категории могут быть достаточно важны при исследовании возраста пользователей
- У женщин, в отличие от мужчин, есть сильное превалирование категории телепередач над всеми остальными. У мужчин нет такого сильного скачка в распределении.
- Просмотров у женщин почти в 2 раза больше, чем у мунчин
- Имеем заметное превалирование возрастной категории 25 - 40
- Чаще всего смотрят видео вечером, абсолютный максимум наблюдается вечером воскресенья, также в воскресенье в разрезе дней недели смотрят чаще других дней. 
- В среднем чаще других видео смотрят пользователи в возрасте от 20 до 30 лет.
- При этом самая многочисленная категория - 30-40 лет
- Помимо телепередач в основном пользователи смотрят сериалы
- Больше всего видео в категории "разное", потом телепередачи и потом сериалы
- Чаще всего пользователи для просмотра контента используют Яндекс.Браузер, Chrome или приложение Rutube лет
